```python
import re
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.preprocessing import LabelEncoder
from collections import defaultdict
from nltk.corpus import wordnet as wn
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import model_selection, naive_bayes, svm
from sklearn.metrics import accuracy_score
```

```python

df = messages = pd.read_csv("hl7_data_small.csv")

df.head()
```

```python
X = df["message"]
y = df["ecg"]
```

```python
messages['message'][1]
```

```python
messages.reset_index(inplace=True)
```

```python
corpus = []

for i in range(0, len(messages)):
    # print(i)
    review = re.sub('[^a-zA-Z0-9]', ' ', messages['message'][i])
    review = re.sub(' +', ' ', review)
    review = review.lower()
    corpus.append(review)

print(i)

corpus[0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
Encoder = LabelEncoder()
Train_Y = Encoder.fit_transform(y_train)
Test_Y = Encoder.fit_transform(y_test)

Train_X = X_train
Test_X = X_test

Train_Y[0:10]
```

```python
Tfidf_vect = TfidfVectorizer(max_features=5000)
Tfidf_vect.fit(corpus)
Train_X_Tfidf = Tfidf_vect.transform(Train_X)
Test_X_Tfidf = Tfidf_vect.transform(Test_X)
```

```python
# fit the training dataset on the NB classifier
Naive = naive_bayes.MultinomialNB()
Naive.fit(Train_X_Tfidf,Train_Y)
# predict the labels on validation dataset
predictions_NB = Naive.predict(Test_X_Tfidf)
# Use accuracy_score function to get the accuracy
print("Naive Bayes Accuracy Score -> ",accuracy_score(predictions_NB, Test_Y)*100)
```

```python
#Corrects some assumptions of Multinomial NB
C_Naive = naive_bayes.ComplementNB()
C_Naive.fit(Train_X_Tfidf,Train_Y)

predictions_CNB = C_Naive.predict(Test_X_Tfidf)

print("Complement Naive Bayes Accuracy Score -> ",accuracy_score(predictions_CNB, Test_Y)*100)
```

```python
B_Naive = naive_bayes.BernoulliNB()
B_Naive.fit(Train_X_Tfidf,Train_Y)

predictions_BNB = B_Naive.predict(Test_X_Tfidf)

print("Bernoulli Naive Bayes Accuracy Score -> ",accuracy_score(predictions_BNB, Test_Y)*100)
```

```python
G_Naive = naive_bayes.GaussianNB()
G_Naive.fit(Train_X_Tfidf.toarray(),Train_Y)

predictions_GNB = G_Naive.predict(Test_X_Tfidf.toarray())

print("Gaussian Naive Bayes Accuracy Score -> ",accuracy_score(predictions_GNB, Test_Y)*100)
```
