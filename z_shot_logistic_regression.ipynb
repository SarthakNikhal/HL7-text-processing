```python
import re
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python

df = messages = pd.read_csv("hl7_data_small.csv")

df.head()
```

```python
X = df["message"]
y = df["ecg"]
```

```python
messages['message'][1]
```

```python
messages.reset_index(inplace=True)
```

```python
corpus = []

for i in range(0, len(messages)):
    # print(i)
    review = re.sub('[^a-zA-Z0-9]', ' ', messages['message'][i])
    review = re.sub(' +', ' ', review)
    review = review.lower()
    corpus.append(review)

print(i)

corpus[0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
Encoder = LabelEncoder()
Train_Y = Encoder.fit_transform(y_train)
Test_Y = Encoder.fit_transform(y_test)

Train_X = X_train
Test_X = X_test

Train_Y[0:10]
```

```python
Tfidf_vect = TfidfVectorizer(max_features=5000)
Tfidf_vect.fit(corpus)
Train_X_Tfidf = Tfidf_vect.transform(Train_X)
Test_X_Tfidf = Tfidf_vect.transform(Test_X)
```

```python
model = LogisticRegression()

model = model.fit(Train_X_Tfidf, Train_Y)

predicted_value = model.predict(Test_X_Tfidf)

print("Logistic Regression Accuracy Score -> ", accuracy_score(predicted_value, Test_Y)*100)
```
